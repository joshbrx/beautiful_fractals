# Sierpinski triangle



---




```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def create_fractal(Points, scale=1/2, init=[0,0], n=10):
    G = np.empty([n + 1, 2])
    X = np.array(init)
    G[0] = X
    temp_list = np.random.choice(np.arange(0, len(Points)), n)
    for i in range(len(temp_list)):
        X = scale * (X + Points[temp_list[i]])
        G[i + 1] = X
    return G
```

```python
Points = np.array([[0,0], [0.5*1.2,1.2*np.sin(np.pi/3)], [1.2,0]])

fractal_pts = create_fractal(Points, n=100000, scale=1/2, init=(1,1))
X, Y = fractal_pts[:,0], fractal_pts[:,1]

plt.figure(figsize=(6,6), dpi=100); # plt.grid()
plt.scatter(X, Y, s=0.01, color="Orange")
for point in Points:
    plt.scatter(point[0], point[1])
# plt.axis('square'); 
plt.show()
```
